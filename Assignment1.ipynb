```python
df['Car'].fillna(df.Car.mode()[0], inplace=True)
df['BuildingArea'].fillna(df.BuildingArea.mean(), inplace=True)
df['YearBuilt'].fillna(df.YearBuilt.median(), inplace=True)
df['CouncilArea'].fillna(df.CouncilArea.mode()[0], inplace=True)
```

```python
df.info()
```

```python
df['YearBuilt'] = df['YearBuilt'].astype(int)
df['Date'] = df["Date"].astype("datetime64")
df['Postcode'] = df['Postcode'].astype('int64')
df['Bedroom2'] = df['Bedroom2'].astype('int64')
df['Bathroom'] = df['Bathroom'].astype('int64')
df['Car'] = df['Car'].astype('int64')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
df['Type'] = encoder.fit_transform(df['Type'])
```

```python
df.Method.unique()
```

```python
df['Method'] = encoder.fit_transform(df['Method'])
```

```python
df['SellerG'] = encoder.fit_transform(df['SellerG'])
df['Regionname'] = encoder.fit_transform(df['Regionname'])
df['CouncilArea'] = encoder.fit_transform(df['CouncilArea'])
```
